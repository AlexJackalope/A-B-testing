```python
import pandas as pd
```

```python
# https://rstudio-pubs-static.s3.amazonaws.com/986404_4dd1a51c486b4c30955a341d938b07ac.html
data = pd.read_csv("./data_viz_website.csv")
data
```

```python
data.condition.unique()
```

```python
import seaborn as sns
```

```python
from matplotlib import pyplot as plt
```

```python
sns.kdeplot(data[data.condition == 'tools'].time_spent_homepage_sec, label='A') 
sns.kdeplot(data[data.condition == 'tips'].time_spent_homepage_sec, label='B')
plt.legend()
```

```python
data.groupby("condition").time_spent_homepage_sec.agg(["count", "median", "max", "mean", "var", "std"])
```

```python
sns.kdeplot(data[data.condition == 'tools'].time_spent_homepage_sec, clip=(49.8, 50.2), color='b', label='A') 
sns.kdeplot(data[data.condition == 'tips'].time_spent_homepage_sec, clip=(49.8, 50.2), color='orange', label='B')
plt.axvline(x=data[data.condition == 'tools'].time_spent_homepage_sec.mean(), color='b')
plt.axvline(x=data[data.condition == 'tips'].time_spent_homepage_sec.mean(), color='orange')
plt.legend()
```
